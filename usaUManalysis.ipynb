```python
import pandas as pd
```

```python
#import seaborn for visualization
import seaborn as sns
```

```python
# importing the dataset
df = pd.read_csv(r"TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
#thinning data(det the exact data we require from here)
# USA athletes
# year 2022
# race 50km and 50 mi

# df.head(10)
usadf = df[(df['Athlete country'] == 'USA') & (df['Year of event']==2022) & df['Event distance/length'].isin(['50km','50mi'])]
```

```python
usadf.info()
```

```python
# reseting the index to have consistent indexing of data
usadf.reset_index(drop=True, inplace=True)
```

```python
usadf.head(10)
```

```python
# 
# 1. We dont need the Athlete country column since we already know the data is based on USA athletes
usadf.drop(columns='Athlete country', inplace=True)
```

```python
# As we can see the Athlete country has been droped
usadf.head(10)
```

```python
# We first have to add the important columns we require
# 1. athlete_age

usadf['athlete_age'] = (usadf['Year of event'] - usadf['Athlete year of birth'])
```

```python
usadf.head()
```

```python
usadf.dropna()
```

```python
# We also  have to drop the
# - Year of event(we already know the data is based in 2022)
# - Athlete club
# - Athlete age category
# - Athlete year of birth
usadf.drop(columns=['Year of event','Athlete year of birth','Athlete age category','Athlete club'], inplace = True)
```

```python
usadf.head(10)
```

```python
usadf.info()
```

```python
# We now cheeck for null value and duplicates to remove them
usadf[usadf.duplicated()]
# here we see there are no duplicates in the dataset

```

```python
# Now we check for null value
usadf.isnull().sum()
# There are null values in the athletes age so we are going to dro them
```

```python
# Drop null values
usadf = usadf.dropna()
```

```python
usadf.info()
```

```python
# Rename the column names for easier accessibility and to avoid typing errors

usadf = usadf.rename (columns=
     {'Event dates': 'event_date',
        'Event name': 'event_name',
        'Event distance/length': 'event_distance',
        'Event number of finishers': 'event_finishers',
        'Athlete performance': 'athlete_performance',
        'Athlete gender': 'gender',
        'Athlete average speed': 'average_speed',
        'Athlete ID': 'athlete_id',
        'athlete_age': 'age'

})
```

```python
usadf.head(10)
# Here wecan see we have renamed the columns
```

```python
# Rearrange the columns(this is a personal preference not necessary)

usadf2 = usadf[['athlete_id','event_date','event_name','event_distance','event_fifnishers','athlete_performance','genger','average_speed','age']]
```

```python
usadf2.head(10)
```

```python
usadf2 = usadf2.rename (columns=
     {'Event dates': 'event_date',
        'event_name': 'event_name',
        'event_distance': 'event_distance',
        'event_fifnishers': 'event_finishers',
        'Athlete performance': 'athlete_performance',
        'genger': 'gender',
        'Athlete average speed': 'average_speed',
        'athlete_id': 'athlete_id',
        'age': 'age'

})
```

```python
usadf2.head(10)
```

```python
# Remove USA from  the event_name column
usadf2.event_name = usadf2.event_name.str.split('(').str.get(0)
```

```python
usadf2.info()
```

```python
# Change age to int type

usadf2.age = usadf2.age.astype(int)
```

```python
# Change average speed to int type
usadf2.average_speed = usadf2.average_speed.astype(float)
```

```python
usadf2.info()
```
